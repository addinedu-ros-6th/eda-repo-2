```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from decimal import Decimal

#출처: kosis 국가통계포털
```

```python
gdpSector = pd.read_csv("../data/GDP_data/gdpSector.csv", encoding="EUC-KR")
gdpSector = gdpSector.set_index(['시도별', '경제활동별', '항목'])
gdpSector.drop('실질기여도', level=2, axis= 0, inplace= True)

# gdpSector = gdpSector.replace(np.nan, 0)
# gdpSector = gdpSector.astype('int')

gdpSector.info()
```

```python
yoy_rate = gdpSector.astype("float")

for i in range(len(yoy_rate.columns)-1):
    yoy_rate.rename(columns= {str(int(i)+2008)+" 년" : (i+2008)}, inplace=True)
    for j in range(len(gdpSector[str(i+2007)+" 년"])):
        if gdpSector.iloc[j,i] != np.nan:
            yoy_rate[i+2008] = gdpSector[str(i+2008)+" 년"]/gdpSector[str(i+2007)+" 년"] -1 

yoy_rate.drop(['2007 년'], axis = 1)

yoy_rate.info()
```

```python
yoy_rate = yoy_rate.reset_index()
yoy_rate = yoy_rate.drop(columns = ['2007 년', 2008])
yoy_rate.info()
```

```python
colins_sort = yoy_rate.loc[yoy_rate['경제활동별'].isin(['농업, 임업 및 어업','\u3000제조업',
                           '서비스업','지역내총생산(시장가격)'])]
colins_sort.replace('제조업', '제조업')
```

```python
colins_sort = colins_sort.replace({'경제활동별' : {'지역내총생산(시장가격)':'총생산', '농업, 임업 및 어업' : '1차 산업', 
                               '\u3000제조업' : '2차 산업', '서비스업' : '3차 산업'}})
# colins_sort = colins_sort.drop(columns= ['level_0', 'index'])
colins_sort.info()
```
